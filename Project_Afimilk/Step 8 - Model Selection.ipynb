# Model Selection

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn import preprocessing
import sklearn.metrics as metrics 
```

```python
df_train = pd.read_pickle(r'C:\DataScience\PROJECT_AFIMILK\Data\Flat_File_for_cleansing\df_train_80.p')
df_val = pd.read_pickle(r'C:\DataScience\PROJECT_AFIMILK\Data\Flat_File_for_cleansing\df_validation_80.p')
```

```python
df_train.head()
```

```python
df_train = df_train.drop(columns=['FarmCode_32','split'])
```

```python
df_val.head()
```

```python
df_val = df_val.drop(columns=['FarmCode_32','split'])
```

```python
X_train = df_train.drop(columns= 'CurMet_t7_t30')
y_train = df_train['CurMet_t7_t30'].astype('category')

print(X_train.shape)
print(y_train.shape)
```

```python
X_val = df_val.drop(columns= 'CurMet_t7_t30')
y_val = df_val['CurMet_t7_t30'].astype('category')

print(X_val.shape)
print(y_val.shape)
```

```python
## Normalized the variables
scaler = preprocessing.StandardScaler().fit(X_train)
X_scaled = scaler.transform(X_train)
##
scaler_val = preprocessing.StandardScaler().fit(X_val)
X_scaled_val = scaler.transform(X_val)
```

```python
def classificationMetrics(y, yhat):
    prf1 = metrics.precision_recall_fscore_support(y,yhat)
    res = {'Accuracy': metrics.accuracy_score(y,yhat),
           'Precision':prf1[0][1],
           'Recall': prf1[1][1],
           'f1-score': prf1[2][1],
           'Log-loss': metrics.log_loss(y,yhat),
           'AUC': metrics.roc_auc_score(y,yhat)
          }
    return res

models_list = pd.DataFrame()
```

## Logistic Regration

```python
mod1 = LogisticRegression(max_iter= 1000, random_state= 1)
mod1.fit(X_scaled,y_train)
```

```python
pred_train1 = mod1.predict(X_scaled)
pd.crosstab(y_train,pred_train1, colnames= ['pred_rtain1'])
```

```python
pred_val1 = mod1.predict(X_scaled_val)
pd.crosstab(y_val, pred_val1, colnames= ['pred_val1'])
```

```python
model_dict = {'model': "Logistic Regression"}
dataset_dict = {'dataset': "Train"}
models_list = models_list.append({**dataset_dict, **model_dict, **classificationMetrics(y_val, pred_val1)}, ignore_index=True)
##
model_dict = {'model': "Logistic Regression"}
dataset_dict = {'dataset': "Validation"}
models_list = models_list.append({**dataset_dict, **model_dict, **classificationMetrics(y_train, pred_train1)}, ignore_index=True)
##
models_list
```
