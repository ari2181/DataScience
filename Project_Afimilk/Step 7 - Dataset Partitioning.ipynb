```python
cat_list = ['Term'
    ,'IsPreviousLactation'
    ,'FarmCode_32'
    ,'LactationNumber_2'
    ,'PPUD_0.0'
    ,'PPUD_1.0'
    ,'DailyRestBout_t3_cat_9.0'
    ,'LastMF_t1_t30i_0.0'
    ,'CurMet_t7_t30'
    ]

for col in cat_list:
    df[col] = df[col].astype('category')
    
obg_list = ['DIMi', 'CurrentDryDaysi']
for col in obg_list:
    df[col] = df[col].astype('float64')
    
```

```python
df_pretrain, df_test = train_test(data=df, prop=0.8, seed=6, tableone=False)
```

```python
print(df_pretrain.shape, df_test.shape)
```

```python
df_train, df_val = train_test(data=df_pretrain, prop=0.8, seed=9, tableone=False)
```
