```python
cat_list = ['Term'
    ,'FarmCode_32'
    ,'PPUD_0.0'
    ,'DailyRestBout_t3_cat_9.0'
    ,'CurMet_t7_t30'
    ]

for col in cat_list:
    df[col] = df[col].astype('category')
    
obg_list = ['DIMi']
for col in obg_list:
    df[col] = df[col].astype('float64')
    
```

```python
df_pretrain, df_test = train_test(data=df, prop=0.8, seed=6, tableone=False)
```
