# Final Test

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_recall_fscore_support, roc_auc_score 
from sklearn import preprocessing
import pickle
```

```python
df_train = pd.read_pickle(r'C:\DataScience\PROJECT_AFIMILK\Data\Flat_File_for_cleansing\df_train_80.p')
df_test = pd.read_pickle(r'C:\DataScience\PROJECT_AFIMILK\Data\Flat_File_for_cleansing\df_test_80.p')
```

```python
df_train
```

```python
df_train = df_train.drop(columns=['FarmCode_32','split'])
```

```python
df_test
```

```python
df_test = df_test.drop(columns=['FarmCode_32','split'])
```

```python
X_train = df_train.drop(columns= 'CurMet_t7_t30')
y_train = df_train['CurMet_t7_t30'].astype('category')

print(X_train.shape)
print(y_train.shape)
```

```python
X_test = df_test.drop(columns= 'CurMet_t7_t30')
y_test = df_test['CurMet_t7_t30'].astype('category')

print(X_test.shape)
print(y_test.shape)
```

```python
## Normalized the variables
scaler = preprocessing.StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
##
scaler_test = preprocessing.StandardScaler().fit(X_test)
X_test_scaled = scaler_test.transform(X_test)
```

```python
def evaluate(model, test_features, test_labels):
    y_pred = model.predict(test_features)
    prf1 = precision_recall_fscore_support(test_labels,y_pred)
    recall = prf1[1][1]
    auc = roc_auc_score(test_labels, y_pred)
    ##
    print('Model Performance')
    print('Recall: {:0.4f}'.format(recall))
    print('AUC: {:0.4f}'.format(auc))
    return recall, auc
```

### ADABoost

```python
filename = r'C:\DataScience\PROJECT_AFIMILK\Data\Flat_File_for_cleansing\adaboost_final_model.p'
model_ada = pickle.load(open(filename, 'rb'))
accuracy_ada = evaluate(model_ada, X_test_scaled, y_test)
```

```python
pred_ada = model_ada.predict(X_test_scaled)
pd.crosstab(y_test, pred_ada, colnames= ['pred_ada'])
```

### Random Forest

```python
filename = r'C:\DataScience\PROJECT_AFIMILK\Data\Flat_File_for_cleansing\rf_final_model.p'
model_rf = pickle.load(open(filename, 'rb'))
accuracy_rf = evaluate(model_rf, X_test_scaled, y_test)
```

```python
pred_rf = model_rf.predict(X_test_scaled) 
pd.crosstab(y_test, pred_rf, colnames= ['pred_rf'])
```

### Logistic Regression

```python
filename = r'C:\DataScience\PROJECT_AFIMILK\Data\Flat_File_for_cleansing\lr_final_model.p'
model_lr = pickle.load(open(filename, 'rb'))
accuracy_lr = evaluate(model_lr, X_test_scaled, y_test)
```

```python
pred_lr = model_lr.predict(X_test_scaled) 
pd.crosstab(y_test, pred_lr, colnames= ['pred_lr'])
```

## Voting

```python
y_vote = pd.DataFrame(index=y_test)
y_vote['ada'] = pred_ada
y_vote['rf'] = pred_rf
y_vote['lr'] = pred_lr
y_vote['sum'] = y_vote['ada'] + y_vote['rf'] + y_vote['lr']
y_vote['vote'] = y_vote['sum'].apply(lambda x : 1 if x >= 1 else 0)
```

```python
print('Model Performance')
print('Recall: ', precision_recall_fscore_support(y_test, y_vote['vote'])[1][1])
print('AUC: ', roc_auc_score(y_test, y_vote['vote']))
```

```python
pd.crosstab(y_test, np.array(y_vote['vote']), colnames= ['y_vote'])
```
