### Multivariable Analysis

```python
y = df['CurMet_t7_t30']
X = df.drop(columns= 'CurMet_t7_t30')
print(X.shape, y.shape)
```

```python
var_selection = var_selection[var_selection['Var'] != 'CurMet_t7_t30']
```

#### Variable Selection using LASSO (L1 penalization)

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

lassomod = Lasso(alpha=0.01).fit(X, y)
```

```python
model = SelectFromModel(lassomod, prefit=True)
model.get_support()
```

```python
var_selection['Lasso'] = model.get_support().astype('int64')
var_selection
```

#### Variable Selection using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

rfmod = RandomForestClassifier().fit(X, y)
```

```python
model = SelectFromModel(rfmod, prefit=True)
model.get_support()
```

```python
var_selection['RandomForest'] = model.get_support().astype('int64')
var_selection
```

#### Variable Selection using Gradient Boosting classification

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.feature_selection import SelectFromModel

gbmod = GradientBoostingClassifier().fit(X, y)
```

```python
model = SelectFromModel(gbmod, prefit=True)
model.get_support()
```

```python
var_selection['GradientBoost'] = model.get_support().astype('int64')
var_selection
```

#### Variable Selection using SVM classification

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel

svmmod = LinearSVC(C=0.01, penalty="l1",dual=False).fit(X, y)
```

```python
model = SelectFromModel(svmmod, prefit=True)
model.get_support()
```

```python
var_selection['SVM'] = model.get_support().astype('int64')
var_selection
```
